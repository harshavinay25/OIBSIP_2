```python
data.columns= ["States","Date","Frequency","Estimated Unemployment Rate","Estimated Employed","Estimated Labour Participation Rate","Region","longitude","latitude"]   
```

DATA VIZUALIZATION

```python
plt.figure(figsize=(5,5))
sns.heatmap(data.select_dtypes(include=[np.number]).corr())
plt.show()
```

PLOTTING BAR GRAPH FOR ESTIMAED UNEMPLOYMENT AND ESTIMAED EMPLOYMENT RATE

```python
data = pd.DataFrame(data)
y=data['Estimated Unemployment Rate']
x=data['States']
pr= plt.figure(figsize=(35, 10))
plt.xlabel("States",fontsize = 40)
plt.ylabel("Estimated Unemployment rate",fontsize = 35)
plt.xticks(fontweight='bold',rotation='vertical',fontsize=20)
plt.bar(x,y, color='teal',align='center')
plt.show()
```

```python
data = pd.DataFrame(data)
y=data['Estimated Employed']
x=data['States']
pr= plt.figure(figsize=(35, 10))
plt.xlabel("States",fontsize = 40)
plt.ylabel("Estimated Employment rate",fontsize = 35)
plt.xticks(fontweight='bold',rotation='vertical',fontsize=20)
plt.bar(x,y, color='#ef9b20',align='center')
plt.show()
```

VIZUALING USING HISTOGRAM GRAPH

```python
plt.figure(figsize=(10, 10))
plt.title("indian unemployment")
sns.histplot(x="Estimated Unemployment Rate",hue = "States",data=data)
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.title("indian employment")
sns.histplot(x="Estimated Labour Participation Rate",hue = "States",data=data)
plt.show()
```

```python
unemploment = data[["States","Region" ,"Estimated Unemployment Rate"]]
figure = px.sunburst(unemploment, path=["Region","States"],values="Estimated Unemployment Rate",width=700, height=700, color_continuous_scale="RdY1Gn",title="Unemployment Rate in India")
figure.show()
```
